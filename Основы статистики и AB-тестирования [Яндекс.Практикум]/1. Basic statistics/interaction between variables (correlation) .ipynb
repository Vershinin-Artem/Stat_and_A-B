- **Коэффициент корреляции Пирсона можно интерпретировать как меру силы линейной зависимости между двумя переменными**
- **Значение коэффициента корреляции Пирсона всегда находится в диапазоне от −1 до 1**

- Большое абсолютное значение коэффициента корреляции Пирсона гарантирует наличие сильной линейной зависимости между переменными
