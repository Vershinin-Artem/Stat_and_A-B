## Задача 3

Выберите верные утверждения. 

- **Если распределение данных скошено, использование медианы обычно более информативно, чем использование среднего значения**
- Медиана всегда лучше, чем среднее, для описания центральной тенденции в данных
- Среднее значение является более устойчивым к выбросам по сравнению с медианой
- **Иногда честнее использовать среднее значение, так как оно учитывает все наблюдения**
- **Выбор между средним значением и медианой зависит от контекста и данных**
